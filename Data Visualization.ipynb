# Line Plot

```python
x = [2010,2011,2012,2013,2014,2015]
y = [50,60,65,55,45,60]
```

```python
plt.plot(x,y,c='r')
```

```python
plt.plot(x,y,c='r');#remove 1st line
plt.xlabel("Year")
plt.ylabel("Sale")
```

```python
y1 = [40,45,60,65,70,60]
```

```python
plt.plot(x,y,c='r');#remove 1st line
plt.plot(x,y1,c='g');
plt.xlabel("Year")
plt.ylabel("Sale")
plt.legend(['Apple','Mango'])
```

# Add market in line plot

```python
plt.plot(x,y,c='r',marker="o");#remove 1st line
plt.plot(x,y1,c='g',marker="o");
plt.xlabel("Year")
plt.ylabel("Sale")
plt.legend(['Apple','Mango'])
```

# fmt shortform [marker,line,color]
    'o--r'  = marker='o',line='--',color='r'

```python
plt.plot(x,y,'o--r');#remove 1st line
plt.plot(x,y1,'*-g');
plt.xlabel("Year")
plt.ylabel("Sale")
plt.legend(['Apple','Mango'])
```

```python
plt.plot(x,y,'or');#remove 1st line
plt.plot(x,y1,'*-g');
plt.xlabel("Year")
plt.ylabel("Sale")
plt.legend(['Apple','Mango'])
```

```python
plt.figure(figsize=(12,6))
plt.plot(x,y,'or');#remove 1st line
plt.plot(x,y1,'*-g');
plt.xlabel("Year",size='20')
plt.ylabel("Sale",size='20')
plt.legend(['Apple','Mango'])
```

```python
df.columns
```

```python
plt.hist(df.sepal_width);
```

```python
#hist use with bins
val = [2,2.5,3,3.5,4,4.5]
plt.hist(df.sepal_width,bins=val);
```

```python
#check unique value in species
list(df['species'].unique())
```

```python
setosa = df[df['species']=='setosa']
versicolor=df[df['species']=='versicolor']
virginica=df[df['species']=='virginica']
```

```python
# plot 2 hist
```

```python
plt.hist(setosa['sepal_width'],alpha=0.4, bins=np.arange(2,5,0.25));
plt.hist(versicolor['sepal_width'],alpha=0.4,bins=np.arange(2,5,0.25));
plt.legend(['setosa','versicolor'])
```

```python
# Stackked histogram
```

```python
plt.hist([setosa['sepal_width'],versicolor['sepal_width'],virginica['sepal_width']],bins=np.arange(2,5,0.25),stacked=True);
plt.legend(['setosa','versicolor','virginica']);
```

# BAR CHART

```python
x=[2000,2001,2002,2003,2004,2005]
y = [10,14,12,16,9,13]
y1 = [12,16,15,13,17,18]
```

```python
df1 = sns.load_dataset('tips')
```

```python
df1.head()
```

```python
plt.bar(df1['day'],df1['total_bill']);
```

```python
avg_total_bill = []
for i in list(df1['day'].unique()):
    avg_total_bill.append(round(np.average(df1[df1['day']==i]['total_bill']),2))
avg_total_bill
```

```python
plt.bar(df1['day'].unique(),avg_total_bill);
plt.title("Avg Total bill of day");
```

# Bar plot and average plot using of seboarn

```python
sns.barplot('day','total_bill',data=df1);
```

```python
sns.barplot('day','total_bill',hue='sex',data=df1);
```

# Matplotlib Markers

```python
y=[1,5,7,3,10,16,20,0,10]
x = np.array(list(range(len(y))))
plt.plot(x,y,marker='o')
plt.show()
```

```python
# ms = markersize 
# mec = markeredgecolor 
# mfc = markerfacecolor 


y=np.array([1,5,7,3,10,16,20,0])
plt.plot(y,'o--',ms=10,mec = 'black',mfc='pink')

plt.show()
```

# Matplotlib Line


```python

y = np.array([5,8,2,3,10,15,11])
plt.plot(y,ls="-")
plt.show()
```

```python

y = np.array([5,8,2,3,10,15,11])
plt.plot(y,ls="-",c='green')
plt.show()
```

```python

y = np.array([5,8,2,3,10,15,11])
plt.plot(y,ls="-",c='green',lw=5.5)
plt.show()
```

```python
y = np.array([1,5,10,2,3,15,8,12])
y2=np.array([2,8,9,14,11,1,7,10])

plt.plot(y)
plt.plot(y2)
plt.show()
```

```python
y = np.array([1,5,10,2,3,15,8,12])
y2=np.array([2,8,9,14,11,1,7,10])

plt.plot(y,lw=3,c='darkgreen',marker='o',mfc='hotpink',ms='10')
plt.plot(y2,ls='-.',c='red',marker='+',ms='15',lw='3',mec='black')
plt.show()
```

```python
y = np.array([1,5,10,2,3,15,8,12])
y2=np.array([2,8,9,14,11,1,7,10])
x = np.array(range(len(y)))
x2=np.array(range(0,16,2))
plt.plot(x,y)
plt.plot(x2,y2)
plt.show()
print(x2)
```

# Matplotlib Labels

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data",)
plt.xlabel("\nAverage Pulse")
plt.ylabel("Calorie Burnage\n")

plt.plot(x, y)
plt.show()
```

```python
# fontdict 
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

xlcolor={'color':'blue','size':20}
ylcolor={'color':'green','size':20}

plt.title("Sports Watch Data",fontdict=ylcolor)
plt.xlabel("\nAverage Pulse",fontdict=xlcolor)
plt.ylabel("Calorie Burnage\n",fontdict=xlcolor)

plt.plot(x,y)
plt.show()
```
